```python
import re # re.search etc. regular expressions

def categorize(description):
    #
    ############################### CHECKING ACCOUNT PROCESSING ###########################
    #
    # Morgage WFHM Wells Fargo Home Morgage
    if 'WFHM' in description.upper():
        return get_category('Morgage')
    # Medical Insurance
    #   VSP Vision Care
    #   ROCKWELL
    elif    'MEDICARE' in description.upper()           or 'VSP' in description.upper() \
            or 'UNITEDHEALTHCARE' in description.upper() \
            or 'ROCKWELL ' in description.upper()       or 'AARP HEALTH' in description.upper() :
        return  get_category('Medical Insurance') 
    # Car Insurance
    elif 'LIBERTY MUTUAL' in description.upper():
        return get_category('Car Insurance')
    # Utilities n Bills
    #   CWD Cleveland Dept of Water
    #   ENBRIDGE GAS
    #   CLEVELAND HEIGHTS sewage ???
    #   NEORSD Northeast Ohio Sewer District
    elif    'VERIZON' in description.upper()                or 'DOMINION ' in description.upper()\
            or 'FIRST ENERGY' in description.upper()        or 'NORTHEAST OHIO' in description.upper() \
            or 'CLEVELAND HEIGHTS ' in description.upper()  or 'ENBRIDGE GAS ' in description.upper() \
            or 'ATT ' in description.upper()                or 'NEORSD ' in description.upper() \
            or 'CWD ' in description.upper() :
        return  get_category('Utilities n Bills')
    # Paper check
    elif    'CHECK' in description.upper()      and not 'E-CHECK' in description.upper() :
        return  get_category('Paper Check - Aikido?')
    # Paper check
    elif    'ATM ' in description.upper():
        return  get_category('ATM Wthw n Dept')
    # College Tuition
    elif    'SMARTPAYCIA' in description.upper()           or 'CASHNET' in description.upper():
        return  get_category('College Tuition')
    # Anita income 
    elif    'ZELLE DEP ANITA' in description.upper() :
        return  get_category('Anita Income')
    # Fidelity Transfer
    elif 'FID ' in description.upper():
        return get_category('Fidelity Transfer')    
    # KeyBank cash-back
    elif    'KEY REWARDS' in description.upper() or 'GIFT FROM KEY BANK' in description.upper() :
        return  get_category('KeyBank cash-back')
    # Taxes and Tax refunds
    elif    'TAXREFUND' in description.upper() or 'IRS ' in description.upper() \
            or 'TAX REF' in description.upper() or 'RITA ' in description.upper() :
        return  get_category('Taxes')
    # Visa Payment
    elif 'INTERNET TRF TO CCA' in description.upper():
        return get_category('Visa Payment') 
    #
    ################################### CREDIT CARD PROCESSING ######################################
    #
    # FEES - FOREIGN contains REI and causes a conflict
    elif 'TRANSACTION FEE' in description.upper():
        return get_category('Fees')
    # POLAND 
    elif    re.search(r' POL$', description.upper()) \
            or re.search(r' IL$', description.upper()) \
            or 'PLUSKI' in description.upper() \
            or 'OLSZTYN' in description.upper() :
        return  get_category('Visiting Grandma Ela')
    # VACATION
    elif    'FOLLY' in description.upper()  \
            or re.search(r' SC$', description.upper()) \
            or re.search(r' NC$', description.upper())  or re.search(r' WV$', description.upper()) \
            or 'VIATORTRIPADVISOR' in description.upper() \
            or 'VACASA' in description.upper()      or 'VRBO' in description.upper() :
        return  get_category('Vacation SC')
    # VISITING WANDA
    elif    re.search(r' VA$', description.upper()) \
            or re.search(r' GA$', description.upper()) \
            or  re.search(r' MD$', description.upper()) :
        return  get_category('Visiting Wanda')
    # VISITING EVA
    elif    re.search(r' PA$', description.upper())           or ' TURNPIKE ' in description.upper() \
            or 'PITTSBURGH' in description.upper() :
        return  get_category('Visiting Eva')
    # GROCERY
    elif    'GROCERY' in description.upper()        or 'HEINEN' in description.upper() \
            or 'DAVE' in description.upper()        or 'WHOLE' in description.upper() \
            or 'SODA' in description.upper()        or 'TRADE' in description.upper() :
        return  get_category('Groceries')
    # DINING OUT
    elif    'TAVERN' in description.upper()         or 'TOMMYS' in description.upper() \
            or 'CAFE' in description.upper() \
            or 'WASABI' in description.upper()      or 'PACIFIC' in description.upper() \
            or 'ANATOLIA' in description.upper()    or 'BATUQUI' in description.upper() \
            or 'PHO' in description.upper()         or 'LAKE HOUSE' in description.upper() \
            or 'DEWEY' in description.upper()       or 'BANANA' in description.upper() \
            or 'BANGKOK' in description.upper()     or 'HIBACHI' in description.upper() \
            or 'LOTUS' in description.upper()       or 'SUBSHOPPE' in description.upper() \
            or 'NATURES OASIS' in description.upper()    or 'BRASSICA' in description.upper() \
            or 'RESTAUR' in description.upper()     or 'BUFFALO' in description.upper() :
        return  get_category('Dining Out')
    # NATIONAL GROCERY n FOOD
    elif    'KRAKOW' in description.upper()        or 'NIPA HUT' in description.upper() :
        return  get_category('World Food')\
    # DOORDASH FOOD
    elif    'DOORDASH' in description.upper() :
        return  get_category('Why DOORDASH ?')
    # FAST FOOD
    elif    'LEFTY' in description.upper()          or 'SHAKE SHACK' in description.upper() \
            or 'WENDY' in description.upper()       or 'BUDDA' in description.upper() \
            or 'CILANTRO' in description.upper()    or 'PANERA' in description.upper() \
            or 'CHIPOTLE' in description.upper()    or 'BIBIBOP' in description.upper() \
            or 'ROGERS' in description.upper():
        return  get_category('Fast Food')
    # BEAUTY n HEALTH
    elif    'LADIES' in description.upper()             or 'BATH' in description.upper() \
            or 'SALLY BEAUTY' in description.upper()    or 'AVEDA' in description.upper() \
            or 'LUSH BEACHWOOD' in description.upper() \
            or 'AIKIKAI' in description.upper()         or 'AIKIDO' in description.upper() \
            or 'ATMA' in description.upper()            or 'PADDLE' in description.upper():
        return  get_category('Beauty n Health')
    # MEDICAL n DENTAL
    # WESTERN RESERVE PERIO - Tomek's paradontist  
    elif    'PEDIATRICS' in description.upper()     or 'CLEVELAND CLINIC' in description.upper() \
            or 'METROHEALTH' in description.upper() \
            or 'WESTERN RESERVE PERIO' in description.upper():
        return  get_category('Medical and Dental')
    # PHARMACY 
    elif    'CVS' in description.upper()            or 'WALGREENS' in description.upper():
        return  get_category('Pharmacy')
    # HOUSE n GARDEN  
    elif    'HOME DEPOT' in description.upper()     or 'BREMEC' in description.upper() \
            or 'LANDSCAPE' in description.upper()   or 'STUMP' in description.upper() \
            or 'NATURE CENTER' in description.upper() or 'REFURNISHING' in description.upper() :
        return  get_category('House and Garden')
    # SUBSCRIPTIONS
    elif    'SPOTIFY' in  description.upper()       or 'APPLE' in description.upper() \
            or 'NETFLIX' in description.upper()     or 'AUDIBLE' in description.upper() \
            or 'PEACOCK' in description.upper()     or 'WALL' in description.upper() \
            or 'BITDEFENDER' in description.upper() or 'MICROSOFT' in description.upper() \
            or 'HULU' in description.upper()        or 'NYTIMES' in description.upper():
        return  get_category('Subscription')
    # AMAZON
    elif    'AMAZON' in description.upper()         or 'AMZN' in description.upper():
        return  get_category('Amazon')
    # DEPARTMENT STORES
    elif    'TARGET' in description.upper()         or 'MACY' in description.upper():
        return  get_category('Department Store')
    # CLOTHES
    elif    'REI' in description.upper()                or 'NORDSTROM' in description.upper() \
            or 'DICK' in description.upper()            or 'DSW' in description.upper() \
            or 'AVALON' in description.upper()          or 'MARSHALLS' in description.upper() \
            or 'ANN TAYLOR' in description.upper() \
            or 'AMERICAN EAGLE' in description.upper()  or 'FOOTWEAR' in description.upper():
        return  get_category('Clothes')
    # PETS
    elif    'PET' in description.upper()            or 'CHEWY' in description.upper() \
            or 'HOLLYWOOD FEED' in description.upper() or 'JACKSON GALAXY' in description.upper() \
            or 'VETERINARY' in description.upper()  or 'PET' in description.upper():
        return  get_category('Pets')
    # KINDLE n BOOKS
    elif    'KINDLE' in description.upper()         or 'AUDIOTEKA' in description.upper() \
        or 'LOGANBERRY' in description.upper()      or 'EMPIK' in description.upper() :
        return  get_category('Kindle n Books')
    # SOFTWARE n ACCESSORIES
    elif    'CLIP STUDIO' in description.upper()    or 'ALIEXPRESS' in description.upper():
        return  get_category('Software n Accessories')
    # ENTERTAINMENT n GIFTS
            # MOVIES THEATER COMPUTER GAMES BIRTHDAY CAKES BDAY DINNERS ETC
            # NOTE: CLEVELAND INSTITUTE OF CLEVELAND ==== CIA CINEMATEQUE
            # MUSEUM ===== includes THE CHILDRENS MUSEUM O CLEVELAND OH 
            # GREATER CLEVELAND AQUA ==== CLEVELAND AQUARIUM
            # MANGO MANGO DESSE CLEVELAND ===== mango mango dessert Sof's birthday with Felix
            # DIAMONDS FLOWERS ========= flower shop Coventry
    elif    'VUDU' in description.upper()               or 'FANDANGO' in description.upper() \
            or 'DOBAMA' in description.upper() \
            or 'CMA' in description.upper()             or 'MUSEUM' in description.upper() \
            or 'THEAT' in description.upper()           or 'CLEVELAND PUBLIC' in description.upper() \
            or 'CLEVELAND INSTITUTE OF CLEVELAND' in description.upper() \
            or 'GREATER CLEVELAND AQUA' in description.upper() \
            or 'BANDCAMP' in description.upper()        or 'MOVIE' in description.upper() \
            or 'PRIME VIDEO' in description.upper()     or 'MITCHELL' in description.upper() \
            or 'BORDERLIGHT' in description.upper()     or 'CINEMA' in description.upper() \
            or 'STEAMGAMES' in description.upper()      or 'STEAM PURCHASE' in description.upper() \
            or 'FIDDLEHEAD' in description.upper()      or 'PASSPORT' in description.upper() \
            or 'SWEET FIX' in description.upper()       or 'STONE OVEN' in description.upper() \
            or 'MANGO MANGO DESSE' in description.upper() or 'DIAMONDS FLOWERS' in description.upper() \
            or 'ON THE RISE' in description.upper()     or 'RISING STAR COFFEE' in description.upper() \
            or 'MICHAELS' in description.upper()        or 'BUNDT' in description.upper() \
            or 'ALL CITY CANDY' in description.upper()  or 'UPTOWN MART' in description.upper() \
            or 'ELLIE-MAYS' in description.upper()      or '6 FLAVORS INDIAN' in description.upper() \
            or 'LUXE KITCHEN' in description.upper() :
        return  get_category('Entertainment n Gifts')
    # GAS n CAR MAINTENANCE
    # NOTE: SHAKER QUALITY AUTO CLEVELAND OH
    elif    'SUNOCO' in description.upper()         or 'BP' in description.upper() \
            or 'REPAIR' in description.upper()      or 'AUTO' in description.upper() \
            or 'MARATHON' in description.upper()    or 'SHELL' in description.upper() \
            or 'CIRCLE K' in description.upper()    or 'SHEETZ' in description.upper() \
            or 'GAS' in description.upper() \
            or 'BODY' in description.upper()        or 'QUALITY AUTO' in description.upper()  \
            or 'BUREAU MOTOR VE'  in description.upper() :
        return  get_category('Gas n Car Maintenance')
    # COLEGE 
    elif    'CIA' in description.upper()            or 'BURREN' in description.upper() \
            or 'UTRECHT' in description.upper()     or 'COLLEGE' in description.upper():
        return  get_category('College n Art Supplies')
    # ART SUPPLIES
    elif    'UTRECHT' in description.upper()            or ' ART ' in description.upper() :
        return  get_category('Art Supplies')
    # AIR TRAVEL
    elif    'LOT ' in description.upper()           or 'AMERICAN' in description.upper() \
            or 'EXPEDIA' in description.upper():
        return  get_category('Air Travel')
    # TOYS
    elif    'PLAYMATTERS' in description.upper()    or 'DISNEYSTORE' in description.upper():
        return  get_category('Kids Toys')
    # POLITICAL DONATIONS
    elif    'ACTBLUE' in description.upper():
        return  get_category('Political Donations')
    # PROFESSIONAL CERT n FEES
    elif    'LICENSURE' in description.upper()      or 'LICENSE' in description.upper():
        return  get_category('Professional Fees')
    # PAYMENT
    elif    'PAYMENT' in description.upper():
        return  get_category('PAYMENT')
    # ACCOUNTED 4 ELSEWHERE
    elif    'DELTA DENTAL' in description.upper():
        return  get_category('Accounted Elsewhere')

    #
    ################################# IN BOTH ACCOUNTS ###########################################
    # OTHER 
    else:
        return  get_category('Uncategorized')
```

```python
def categorize_all(input_period):
    credit_acct_suffix = 'CCard'
    checking_acct_suffix = 'CheckingAcct'
    credit_input_datafile = input_period + '-A&T-' + credit_acct_suffix + '.xlsx'
    checking_input_datafile = input_period + '-A&T-' + checking_acct_suffix + '.xlsx'
    output_datafile_summary = input_period + '-A&T-ALLACCOUNTS-SUMMARY.csv'

    # DEBUG DATA files and FLAG
    PRINT_DEBUG_DATA = True

    # Load the credit card data
    credit_card_data = pd.read_excel(credit_input_datafile)
    credit_card_data.columns = ['Date', 'Amount', 'Description', 'Ref,#']
    # Categorize by description and add Category column
    credit_card_data['Category'] = credit_card_data['Description'].apply(categorize)

    # DEBUG DATA files and FLAG
    if PRINT_DEBUG_DATA:
        output_datafile_categorized = input_period + credit_acct_suffix + '-A&T-CATEGORIZED.csv'
        output_datafile_categorized_sorted = input_period + credit_acct_suffix + '-A&T-SORTED-BY-CATEGORY.csv'
        # Save the updated DataFrame to a new CSV file
        credit_card_data.to_csv(output_datafile_categorized, index=False)
        # Group transactions by category (e.g., food, clothing, entertainment) and calculate total amounts for each category
        sorted_by_category = credit_card_data.sort_values(by='Category', ascending=True)
        # save sorted data to a file
        sorted_by_category.to_csv(output_datafile_categorized_sorted)

    # Summarize amounts in grouped categories
    credit_grouped_summary = credit_card_data.groupby('Category')['Amount'].sum().reset_index()
    credit_grouped_summary = credit_grouped_summary.rename(columns={'Amount': 'Credit'})

    # Load the checking card data
    checking_card_data = pd.read_excel(checking_input_datafile)
    checking_card_data.columns = ['Date', 'Amount', 'Description', 'Ref,#']
    # Categorize by description and add Category column
    checking_card_data['Category'] = checking_card_data['Description'].apply(categorize)

    if PRINT_DEBUG_DATA:
        # DEBUG DATA files and FLAG
        output_datafile_categorized = input_period + checking_acct_suffix + '-A&T-CATEGORIZED.csv'
        output_datafile_categorized_sorted = input_period + checking_acct_suffix + '-A&T-SORTED-BY-CATEGORY.csv'
        # Save the updated DataFrame to a new CSV file
        checking_card_data.to_csv(output_datafile_categorized, index=False)
        # Group transactions by category (e.g., food, clothing, entertainment) and calculate total amounts for each category
        checking_sorted_by_category = checking_card_data.sort_values(by='Category', ascending=True)
        # save sorted data to a file
        checking_sorted_by_category.to_csv(output_datafile_categorized_sorted)
        
    # Summarize amounts in grouped categories
    checking_grouped_summary = checking_card_data.groupby('Category')['Amount'].sum().reset_index()
    checking_grouped_summary = checking_grouped_summary.rename(columns={'Amount': 'Checking'})
    
    #cat_num = data["Category"].nunique()
    cat_num = all_categories["Category"].nunique()
    merged_summary = all_categories.merge(credit_grouped_summary, on='Category', how='left')
    merged_summary = merged_summary.merge(checking_grouped_summary, on='Category', how='left')
    merged_summary = merged_summary.fillna(0)

    # add a column summarizing both Amount
    # Assume the columns are named 'Amount_credit' and 'Amount_checking'
    merged_summary['Total'] = merged_summary['Credit'] + merged_summary['Checking'] 
    
    print(merged_summary)

    # sorted_summary = grouped_summary.sort_values(by='Amount', ascending=False).head(cat_num)
    sorted_summary = merged_summary.sort_values(by='Priority', ascending=True).head(cat_num)
    sorted_summary.to_csv(output_datafile_summary, index=False)
```
